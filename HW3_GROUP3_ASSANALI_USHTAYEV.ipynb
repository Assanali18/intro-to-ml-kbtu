Express $\sigma''(x)$ in terms of $\sigma(x)$ where $\sigma(x) = \frac 1{1 + e^{-x}}$  sigmoid function. Find all $x$ for which $\sigma''(x) = 0$. Also find intervals on which sigmoid is convex/concave.

### YOUR SOLUTION HERE

$$\sigma'(x) = \frac {e^{-x}}{(1 + e^{-x})^{2}} \Longrightarrow  \sigma''(x) = \frac {e^{-x}-e^{-2x}}{(1 + e^{-x})^{3}} $$

$$

## Task 3.3 (2 points)

Let $\boldsymbol A \in \mathbb R^{n\times n}$ be a symmetric matrix and

$$
f(\boldsymbol x) = \frac{\boldsymbol x^\mathsf{T} \boldsymbol{Ax}}{\boldsymbol x^\mathsf{T} \boldsymbol x}, \quad \boldsymbol x \in \mathbb R^n.
$$ 

(a) (**1 point**) Calculate $\nabla f(\boldsymbol x)$ and prove that $\nabla f(\boldsymbol x) = \boldsymbol 0$ iff $\boldsymbol x$ is an eigenvector of $\boldsymbol A$.

if A - symmetric $\Longrightarrow$  $A = A^{T}$
Let B = $\boldsymbol x^\mathsf{T} \boldsymbol{Ax}$, C = $\boldsymbol x^\mathsf{T} \boldsymbol x$. 
$$f(\boldsymbol x) = \frac{B}{C}$$

$$\nabla f(\boldsymbol x) = \frac{\nabla BC - B\nabla C}{C^{2}}$$

$$dB = \nabla B^\mathsf{T} dx \Longrightarrow dB = d( \boldsymbol x^\mathsf{T} \boldsymbol{Ax}) = (d \boldsymbol x^\mathsf{T}) \boldsymbol{Ax} + \boldsymbol x^\mathsf{T}d(Ax) \Longrightarrow (d \boldsymbol x^\mathsf{T})(\boldsymbol x^\mathsf{T}A^\mathsf{T})^\mathsf{T} + \boldsymbol x^\mathsf{T}A dx = \boldsymbol x^\mathsf{T}A^\mathsf{T}dx + \boldsymbol x^\mathsf{T}Adx \Longrightarrow \nabla B = 2A\boldsymbol x$$
Similar with C:
$$C = 2\boldsymbol x$$

In result:

$$\nabla f = \frac{2(\boldsymbol x^\mathsf{T}x \cdot Ax - x \cdot x^\mathsf{T}Ax)}{(x^\mathsf{T} \cdot x)^{2}}$$


when x is a eigenvalue $\Longrightarrow  \nabla f = \frac{2(\boldsymbol x^\mathsf{T} \lambda - x^\mathsf{T} \lambda)}{(x^\mathsf{T} \cdot x)^{2}} = 0$



(b) (**1 point**) Prove that if $\boldsymbol A$ is positive definite then

$$
\max\limits_{\boldsymbol x \ne \boldsymbol 0} f(\boldsymbol x) = \lambda_{\max}(\boldsymbol A), \quad
\min\limits_{\boldsymbol x \ne \boldsymbol 0} f(\boldsymbol x) = \lambda_{\min}(\boldsymbol A)
$$



a)To find the maximum value of $f(x)$, we can use the Rayleigh quotient theorem, which states that for a symmetric matrix A with eigenvalues $\lambda_{1} \leq \lambda_{2} ... \leq \lambda_{n}$ and corresponding orthonormal eigenvectors $v_{1},v_{1}, ... v_{n}$, the maximum and minimum values of $\frac{\boldsymbol x^\mathsf{T} \boldsymbol{Ax}}{\boldsymbol x^\mathsf{T} \boldsymbol x}$ over all nonzero vectors $x$ are $\lambda_{max}(A)$ and $\lambda_{min}(A)$ respectively, and these values are attained when  $x$ is an eigenvector corresponding to $\lambda_{max}(A)$ or $\lambda_{min}(A)$ respectively.

Since A is positive definite, all its eigenvalues are positive. Therefore, 
$\lambda_{\max}(A)$ is the largest eigenvalue, and $v_{max}$ is the corresponding eigenvector. Let $x = v_{max}$, then $f(\boldsymbol x) = \frac{\boldsymbol x^\mathsf{T} \boldsymbol{Ax}}{\boldsymbol x^\mathsf{T} \boldsymbol x} = \frac{\lambda_{\max}(A) \boldsymbol x^\mathsf{T} \boldsymbol x}{\boldsymbol x^\mathsf{T} \boldsymbol x} = \lambda_{\max}(A)$.

So, $\max\limits_{\boldsymbol x \ne \boldsymbol 0} f(\boldsymbol x) = \lambda_{\max}(\boldsymbol A)$.


b)Similarly, for the minimum value of $f(x)$, we choose $x = v_{min}$, where $v_{min}$ is the eigenvector corresponding to the smallest eigenvalue $\lambda_{\min}(A)$. Thus, $f(\boldsymbol x) = \frac{\boldsymbol x^\mathsf{T} \boldsymbol{Ax}}{\boldsymbol x^\mathsf{T} \boldsymbol x} = \frac{\lambda_{\min}(A) \boldsymbol x^\mathsf{T} \boldsymbol x}{\boldsymbol x^\mathsf{T} \boldsymbol x} = \lambda_{\min}(A)$.

So, $\min\limits_{\boldsymbol x \ne \boldsymbol 0} f(\boldsymbol x) = \lambda_{\min}(\boldsymbol A)$.
